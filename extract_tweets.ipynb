```python
#tweets_list = tweepy.Cursor(api.search, q="gorda", geocode="-9.189967,-75.015152,2500km", include_rts=False, tweet_mode="extended").items(200)  
#tweets_list = tweepy.Cursor(api.search, q="#GasNatural -filter:retweets",tweet_mode="extended").items(100) 
#tweets_list = tweepy.Cursor(api.search, q="#GasNatural -filter:retweets",since="2020-05-01",until="2021-10-24",lang="es",tweet_mode="extended").items(100) 

output = []
tweets = api.user_timeline(screen_name="ppkamigo", 
                           # 200 is the maximum allowed count
                           count=200,
                           include_rts = False,
                           #Necessary to keep full_text 
                           #otherwise only the first 140 words are extracted
                           tweet_mode = 'extended'
                           )

for info in tweets:
    #print("ID: {}".format(info.id))
    #print(info.created_at)
    #print(info.full_text)
    #print("\n")
    text = info.full_text
    text = re.sub(r"[\r\n]+","",text)
    line = {'id':info.id, 'date':info.created_at ,'tweet' : text}
    output.append(line)
    #print(line)
```
